```python
w_lb=float(input('Weight(in pounds):'))
h_ft=int(input('Height(in feet):'))
h_in=float(input('Remaining height(in inches):'))
w_kg=w_lb//2.205
print(w_kg,'kg')
h_cm=round(((h_ft)*30.48)+((h_in)*2.54),2)
print(h_cm,'cm')
h_m=round(h_cm/100,2)
print(h_m,'m')
bmi=round(w_kg/(h_m)**2,1)
print(bmi)
if bmi<18.5:
  print('Underweight')
else:
  if bmi>=18.5 and bmi<=24.9:
    print('Normal weight')
  if bmi>=25 and bmi<=29.9:
    print('Overweight')
  else:
    if bmi>30:
      print('Obesity')
```
